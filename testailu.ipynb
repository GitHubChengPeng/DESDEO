```python

from desdeo.utopia_stuff.utopia_problem import utopia_problem
from desdeo.mcdm.nimbus import generate_starting_point
from desdeo.tools.utils import available_solvers
```

```python
prob, schedule_dict = utopia_problem()
start_result = generate_starting_point(problem=prob,solver=available_solvers["gurobipy"])
```

```python
for key in start_result.optimal_variables:
    if not key.startswith("X"):
        continue
    print(schedule_dict[key]["unit"])
    print(schedule_dict[key][start_result.optimal_variables[key].index(1)])
```

```python
description_dict = {
    0: "Do nothing",
    1: "Clearcut",
    2: "Thinning from below",
    3: "Thinning from above",
    4: "Even thinning",
    5: "First thinning",
}

def treatment_index(part: str)->str:
    if "clearcut" in part:
        return 1
    if "below" in part:
        return 2
    if "above" in part:
        return 3
    if "even" in part:
        return 4
    if "first" in part:
        return 5
    return -1

treatments_dict={}
for key in start_result.optimal_variables:
    if key.startswith("_"):
        continue
    treatments = schedule_dict[key][start_result.optimal_variables[key].index(1)]
    treatments_dict[key] = {"2025":0, "2030":0, "2035":0}
    for year in treatments_dict[key]:
        if year in treatments:
            print(treatments)
            for part in treatments.split():
                if year in part:
                    treatments_dict[key][year]=treatment_index(part)


```
