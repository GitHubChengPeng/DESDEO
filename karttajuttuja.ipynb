```python
# Define a function to create a schema from a DataFrame
def create_schema(df):
    # Initialize the schema dictionary
    schema = {
        'geometry': 'None',  # No geometry for non-spatial data
        'properties': {}
    }
    
    # Map pandas types to Fiona types
    type_mapping = {
        'int64': 'int',
        'float64': 'float',
        'object': 'str',  # Typically for strings
        'bool': 'bool',
        'datetime64[ms]': 'str'
    }

    # Iterate through DataFrame columns and build the properties
    for column in df.columns:
        # Get the type of the column
        col_type = str(df[column].dtype)
        
        # Map the pandas type to Fiona type and add it to the schema
        if col_type in type_mapping:
            schema['properties'][column] = type_mapping[col_type]
        else:
            raise ValueError(f"Unsupported data type: {col_type} for column: {column}")
    
    return schema
```
